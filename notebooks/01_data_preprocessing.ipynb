```python
from acnportal import acndata
import pandas as pd
import json
```

```python
client = acndata.DataClient(api_token='DEMO_TOKEN')
```

```python
docs = client.get_sessions_by_time(
    site='caltech',
    start='2018-01-01',
    end='2020-12-31',
    limit=10000  # start with 10k sessions
)
```
